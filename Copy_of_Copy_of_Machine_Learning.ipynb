```python
mean_Recommend = df[['Stars', 'calories']][df['Recommend'] == 'Yes'].mean()
mean_notRecommend = df[['Stars', 'calories']][df['Recommend'] == 'NO'].mean()

var_Recommend = df[['Stars', 'calories']][df['Recommend'] == 'Yes'].mean()
var_notRecommend = df[['Stars', 'calories']][df['Recommend'] == 'NO'].mean()
```

```python
print('Mean values for Recommend features\n===\n{}'.format(mean_Recommend))
print()
print('Mean values for notRecommend features\n===\n{}'.format(mean_notRecommend))
print()
print('Variance values for Recommend features\n===\n{}'.format(var_Recommend))
print()
print('Variance values for notRecommend features\n===\n{}'.format(var_notRecommend))
```

```python
def likelihood(feature, mean, variance):
    return (1 / np.sqrt(2 * np.pi * variance)) * np.exp((-(feature - mean) ** 2))
```
